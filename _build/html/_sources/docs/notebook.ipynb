## Packages importeren

## Data inladen

```python
# inladen van de csv bestanden
df1 = pd.read_csv('canadian_immegration_data.csv')
df2 = pd.read_csv('Life Expectancy Data.csv')
df3 = pd.read_csv('GlobalLandTemperaturesByCountry.csv')
```

## 10 rijkste en 10 armste landen op basis van gemiddeld BBP

De bar plots onder deze tekst tonen de economische kracht van landen door de top 10 rijkste en armste landen te visualiseren op basis van hun BBP (Bruto Binnenlands Product) in de periode 2010-2015. De bovenste grafiek vertegenwoordigt de top 10 rijkste landen, terwijl de onderste grafiek de top 10 armste landen weergeeft. De lengte van de bars correspondeert met het BBP van elk land. Deze visualisatie biedt een overzicht van de economische ongelijkheid tussen landen en benadrukt de grote verschillen in economische welvaart over de hele wereld. 

De belangrijkste boodschap die uit deze bar plot naar voren komt, is de economische kloof tussen de top 10 rijkste en armste landen. Het stelt de lezer in staat om de landen te identificeren die de hoogste en laagste BBP-waarden hebben en biedt inzicht in de economische ongelijkheid op wereldschaal gedurende de specifieke periode van 2010-2015. In het vervolg van het project worden deze landen gebruikt om te kijken hoeveel BBP-invloed heeft op de immigratie naar Canada.  

```python
import matplotlib.pyplot as plt

# pakt de mean gdp van alle landen in de df2 dataset
mean_per_country = df2.groupby('Country')['GDP'].mean()

# selecteert de top 10 van mean_per_country
top_10_landen = mean_per_country.nlargest(10)

# selecteert de laagste 10 van mean_per_country
laagste_10_landen = mean_per_country.nsmallest(10)

# maakt een figuur met twee subplots
figuurtje, (plot1, plot2) = plt.subplots(2, 1, figsize=(10, 10))

# maakt de plot met de top_10_landen index en waarden
plot1.bar(top_10_landen.index, top_10_landen.values)

# voegt labels toe an de plot
plot1.set_xlabel('Land')
plot1.set_ylabel('Gemiddeld BBP')

# voegt titel toe aan de plot
plot1.set_title('Rijkste 10 landen op basis van gemiddeld BBP')

# draait de x-as labels 45 graden
plot1.tick_params(axis='x', rotation=45)

# maakt de plot met de laagste_10_landen index en waarden
plot2.bar(laagste_10_landen.index, laagste_10_landen.values)

# voegt labels toe aan de plot
plot2.set_xlabel('Land')
plot2.set_ylabel('Gemiddeld BBP')

# voegt titel toe aan de plot
plot2.set_title('Armste 10 landen op basis van gemiddeld BBP')

# draait de x-as labels 45 graden
plot2.tick_params(axis='x', rotation=45)

# past de layout van de subplots aan
plt.tight_layout()

# laat de plots zien
plt.show()
```

## Totaal aantal immigranten naar Canada op basis van gemiddeld BBP

De bar chart onder deze tekst geeft het aantal immigranten weer die uit een bepaald land naar Canada gaan. De selectie van 20 landen bestaat uit 10 van de landen met het hoogste BBP ter wereld en 10 van de landen met het laagste BBP ter wereld. De landen zijn oplopend van links naar rechts gerangschikt, wat resulteert in een balkgrafiek. De y-as vertegenwoordigt het aantal immigranten van een land. De x-as vertegenwoordigt om welk van de 20 landen het gaat. Elke balk correspondeert met een specifiek land en de hoogte van de balk geeft de immigratie naar Canada vanuit dat land aan. De landen uit de lijst met het hoogste BBP ter wereld hebben een blauwe kleur balk en de landen uit de lijst met het laagste BBP ter wereld hebben een rode kleur balk. Deze twee verschillende kleuren stelt een visuele analyse in staat snel en gemakkelijk het verschil te zien tussen welk soort landen een hoger aantal immigranten hebben die naar Canada vertrekken.

De belangrijkste boodschap die uit deze bar chart naar voren komt, is de variatie in immigratiecijfers van verschillende landen naar Canada. Het stelt de lezer in staat om de landen te identificeren die de hoogste immigratie naar Canada hebben, terwijl ook duidelijk wordt welke landen relatief lagere immigratiecijfers hebben. Deze visualisatie biedt inzicht in de migratiestromen naar Canada en benadrukt de landen waaruit de meeste immigranten afkomstig zijn, wat kan bijdragen aan het begrijpen van demografische patronen en migratietrends.

```python
# selectie van hoogste en laagste 10 landen op basis van gemiddeld BBP
selectie_landen = [x for x in top_10_landen.index]
for x in laagste_10_landen.index:
    selectie_landen.append(x)

# selecteerd de landen in de dataset die in selectie_landen zitten
df_landen = df1[df1['Country'].isin(selectie_landen)]

# sorteert de df_landen op de kolom totaal
df_landen = df_landen.sort_values('Total')

# maakt de traces aan voor de plot
trace = go.Bar(
    x=df_landen['Country'],
    y=df_landen['Total'],
    marker=dict(
        color=['blue' if land in top_10_landen.index else 'red' for land in df_landen['Country']]
    )
)

# zet de titel en de as titels in de variabel opmaak
opmaak = go.Layout(
    title='Totaal aantal immigranten naar Canada op basis van gemiddeld BBP',
    xaxis=dict(title='Land'),
    yaxis=dict(title='Totaal aantal immigranten'),
)

# maakt de plot met alle data zoals opmaak en layout
fig = go.Figure(data=[trace], layout=opmaak)

# laat do plot zien
fig.show()
```

## Sterfte onder volwassenen en totale immigratie

De scatter plot onder deze tekst illustreert de relatie tussen de variabele 'Sterfte onder volwassenen' en de 'Totale immigratie' (naar Canada). Op de x-as wordt het sterftecijfer onder volwassenen (Sterfte onder volwassenen) weergegeven, terwijl op de y-as de totale immigratie (Totale immigratie) naar Canada staat. Elk gekleurd stipje in het plot vertegenwoordigt een specifiek land. De legenda naast de grafiek geeft de corresponderende landen en kleuren aan. 

Deze visualisatie stelt ons in staat om verbanden te onderzoeken tussen het sterftecijfer onder volwassenen en de immigratie naar Canada. Een hoog sterftecijfer kan wijzen op een slechte situatie in een land, veroorzaakt door factoren zoals ontoereikende gezondheidszorg, levensomstandigheden of andere oorzaken. Door de positie van de landen op de plot te analyseren, kunnen we inzicht krijgen in hoe deze factoren mogelijk van invloed zijn op de immigratiepatronen naar Canada. Het onderzoeken van de spreiding van de stipjes en eventuele opvallende punten kan helpen om de relatie tussen adult mortality en immigratie beter te begrijpen. 

```python
# voegt datasets samen op basis van land
samendf = pd.merge(df2, df1, on='Country')

# verwijdert duplicaten in de samengevoegde dataset
samendf.drop_duplicates(subset='Country', inplace=True)

# sorteert aflopend op basis van totale immigratie
samendf.sort_values('Total', ascending=False, inplace=True)

# selecteert hoogste en laagste landen op basis van totale immigratie
hoogste_10_landen2 = samendf.head(10)
laagste_10_landen2 = samendf.tail(10)

# voegt de hoogste en laagste landen samen
samen_extreme_landen = pd.concat([hoogste_10_landen2, laagste_10_landen2])

# geeft elk land een andere kleur
aantal_countries = len(samen_extreme_landen)
kleuren = plt.cm.tab20(np.linspace(0, 1, aantal_countries))

# plot de levensverwachting en totale immigratie per land
fig, assen2 = plt.subplots(figsize=(10, 8))
for i, country in enumerate(samen_extreme_landen.iterrows()):
    assen2.scatter(country[1]['Adult Mortality'], country[1]['Total'], color=kleuren[i], label=country[1]['Country'])

# voegt labels toe aan de plot
plt.xlabel('Sterfte onder volwassenen')
plt.ylabel('Totale immigratie')

# voegt een titel toe aan de plot
plt.title('Sterfte onder volwassenen en totale immigratie')

# maakt een legenda voor de plot buiten de plot
legend = plt.legend(title='Land', loc='center left', bbox_to_anchor=(1, 0.5))
plt.setp(legend.get_title(), fontsize='large')

# laat de plot zien
plt.show()
```

## 10 Warmste en 10 koudste landen

Voor het tweede perspectief wilden we onderzoeken of er een relatie was tussen de temperatuur van een land en de immigranten die naar Canada zijn gegaan. Voor de temperaturen van de landen hebben we een dataset gebruikt die van alle landen vanaf 1750 tot 2020 alle temperaturen bevat. We hebben de gemiddelde temperatuur berekent van elk land tussen 1890 en 2013 en daarvan de tien warmste landen en tien koudste landen in twee grafieken gezet. Dit zijn de eerste twee grafieken. Vervolgens hebben we een grafiek gemaakt van het totaal aantal immigranten die naar Canada zijn gegaan van de tien warmste landen en tien koudste landen. Hier kan je zien dat van de top tien warmste landen de meeste immigranten uit Senegal kwamen en van de tien koudste landen de meeste immigranten uit Denemarken kwamen.  

Hierna hebben we een plot gemaakt met al die twintig landen met op de x-as de gemiddelde temperatuur van de landen en op de y-as het aantal immigranten naar Canada. In deze grafiek kan je goed zien dat het aantal immigranten naar Canada goed is verdeeld over de verschillende temperaturen. Daarom zijn wij tot de conclusie gekomen dat temperatuur niet veel invloed heeft op de immigranten stroom naar Canada. Als we niet de tien warmste en koudste landen hadden genomen en een andere aanpak hadden genomen, waren we misschien op een andere conclusie gekomen. 

```python
# verandert de dt kolom naar datatime data
df3['dt'] = pd.to_datetime(df3['dt'])

# Filtert df3 op alleen data tussen 1890-01-01 en 2013-12-31
begin = pd.to_datetime('1890-01-01')
einde = pd.to_datetime('2013-12-31')
gefilterd = df3[(df3['dt'] >= begin) & (df3['dt'] <= einde)]

# groepeert de data op land en berekent de gemiddelde temperatuur
gem_temp_landen = gefilterd.groupby('Country')['AverageTemperature'].mean()

# selecteert de hoogste 10 landen op basis van gemiddelde temp
hoogste_10_landen_gemtemp = gem_temp_landen.nlargest(10)

# verandert de grootte van de plot
figuur, assen = plt.subplots(figsize=(10, 6))

# maakt de bar plot aan met de data van hoogste_10_landen_gemtemp
bar_plot = assen.bar(hoogste_10_landen_gemtemp.index, hoogste_10_landen_gemtemp.values)

# voegt labels toe aan de plot
assen.set_xlabel('Land')
assen.set_ylabel('Gemiddelde temperatuur (C\N{DEGREE SIGN})')

# voegt een titel toe aan de plot
assen.set_title('Hoogste landen op basis van gemiddelde temperatuur')

# Geeft de tick posities aan (anders kregen we een error)
assen.set_xticks(range(len(hoogste_10_landen_gemtemp.index)))

# draait de x-as labels 45 graden
assen.set_xticklabels(hoogste_10_landen_gemtemp.index, rotation=45, ha='right')

# voegt de temperaturen van de landen toe boven aan de positieve staven en onder negatieve
for rect in bar_plot:
    height = rect.get_height()
    if height >= 0:
        assen.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')
    else:
        assen.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2f}', ha='center', va='top')

# laat de plot zien
plt.tight_layout()
plt.show()

# selecteert de laagste 10 landen op basis van gemiddelde temp
laagste_10_landen_gemtemp = gem_temp_landen.nsmallest(10)

# verandert de grootte van de plot
figuur, assen = plt.subplots(figsize=(10, 6))

# maakt de bar plot aan met de data van laagste_10_landen_gemtemp
bar_plot = assen.bar(laagste_10_landen_gemtemp.index, laagste_10_landen_gemtemp.values)

# voegt labels toe aan de plot
assen.set_xlabel('Land')
assen.set_ylabel('Gemiddelde temperatuur (C\N{DEGREE SIGN})')

# voegt een titel toe aan de plot
assen.set_title('Laagste landen op basis van gemiddelde temperatuur')

# Geeft de tick posities aan (anders kregen we een error)
assen.set_xticks(range(len(laagste_10_landen_gemtemp.index)))

# draait de x-as labels 45 graden
assen.set_xticklabels(laagste_10_landen_gemtemp.index, rotation=45, ha='right')

# voegt de temperaturen van de landen toe boven aan de staven
for rect in bar_plot:
    height = rect.get_height()
    if height >= 0:
        assen.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')
    else:
        assen.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2f}', ha='center', va='top')

# laat de plot zien
plt.tight_layout()
plt.show()

print('Voor het maken van deze twee grafieken is op 20 juni 2023 gebruik gemaakt van ChatGPT (versie GPT3.5). Gebruikte prompts waren "op het moment staan de getallen die negatief zijn in de staven van de grafiek terwijl de temperatuur getallen juist onder aan de staven moet staan." en "in deze grafiek zijn er 3 staven die wel een positieve temperatuur hebben dus hier moet de temperatuur juist boven de staaf staan."')
```

## Gemiddelde temperatuur en totaal aantal immigranten naar Canada

tekst over onderstaande grafiek

```python
# verandert de dt kolom naar datatime data
df3['dt'] = pd.to_datetime(df3['dt'])

# Filtert df3 op alleen data tussen 1890-01-01 en 2013-12-31
gefilterde_data = df3[(df3['dt'] >= begin) & (df3['dt'] <= einde)]

# groepeert de data op land en berekent de gemiddelde temperatuur
gemp_temp_per_land = gefilterde_data.groupby('Country')['AverageTemperature'].mean()

# voegt gem temperatuur en totale immigratie datasets samen op basis van land
samengevoegd_data = pd.merge(gemp_temp_per_land, df1, on='Country')

# sorteert aflopend de data op basis van gemmidelde temperatuur
samengevoegd_data.sort_values('AverageTemperature', ascending=False, inplace=True)

# maakt selectie van top 10 warmste en koudste landen
top_10_warmste_landen = samengevoegd_data.head(10)
top_10_koudste_landen = samengevoegd_data.tail(10)

# voegt de top warmste en laagste koudste samen
selectie_landen = pd.concat([top_10_warmste_landen, top_10_koudste_landen])

# geeft elk land een andere kleur
aantal_landen = len(selectie_landen)
kleuren = plt.cm.tab20(np.linspace(0, 1, aantal_landen))

# plot de gem temp en totale immigratie per land
fig, ax = plt.subplots(figsize=(10, 8))
for i, (_, country) in enumerate(selectie_landen.iterrows()):
    ax.scatter(country['AverageTemperature'], country['Total'], c=[kleuren[i]], label=country['Country'])

# voegt labels toe aan de plot
plt.xlabel('Gemiddelde temperatuur')
plt.ylabel('Totaal aantal immigranten')

# voegt totel toe aan de plot
plt.title('Gemiddelde temperatuur en totaal aantal immigranten naar canada')

# maakt een legenda voor de plot buiten de plot
legend = ax.legend(title='Land', loc='center left', bbox_to_anchor=(1, 0.5))
plt.setp(legend.get_title(), fontsize='large')

# laat de plot zien
plt.show()
```
